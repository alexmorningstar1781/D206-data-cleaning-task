```python
from sklearn.decomposition import PCA
```

```python
#columns to be used in PCA assigned to variable
pca_df=df[['Age','Lat','Lng','Population','Income','VitD_levels','Initial_days','TotalCharge','Additional_charges']]
```

```python
#normalizes the data in the columns and assigns to a new variable
pca_df_norm=(pca_df-pca_df.mean())/pca_df.std()
```

```python
#have python determine the number of PCA components
pca=PCA(n_components=pca_df_norm.shape[1])
```

```python
pca.fit(pca_df_norm)
```

```python
med_pca=pd.DataFrame(pca.transform(pca_df_norm),columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9'])
```

```python
loadings=pd.DataFrame(pca.components_.T, columns=['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9'], index=pca_df.columns)
```

```python
print(loadings)
```

```python
cov_matrix=np.dot(pca_df_norm.T,pca_df_norm)/pca_df.shape[0]
```

```python
eigenvalues=[np.dot(eigenvector.T,np.dot(cov_matrix,eigenvector)) for eigenvector in pca.components_]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(eigenvalues)
plt.title('Scree Plot')
plt.xlabel('num of comps')
plt.ylabel('eigenvalues')
plt.axhline(y=1,color="red")
plt.show()
```
